```python
# Collect TVL and prices of TVL
import csv
n=4 # number of vaults - don't change!!!

with open("vaultstvl51.csv", encoding='utf-8') as r_file:
    
    file_reader = csv.reader(r_file, delimiter = ";")
    counter=0
    price=[]
    tvl=[]
    
    for row in file_reader:
        if counter==0:
            counter = 1
            print('Заголовки = ',row)
        else:
            #print('row = ', row)
            price.append([0]*(n+1))
            tvl.append([0]*(n+1))
            #print('counter= ', counter)
            for i in range(0,n+1,1):
                #print('i=', i)
                if i==0:
                    price[counter-1][0]=row[0]
                    tvl[counter-1][0]=row[0]     
                else:
                    #print('row 2= ', row)
                    #print('i= ',i)
                    price[counter-1][i]=float(row[2*i])#float(row[0])
                    tvl[counter-1][i]=float(row[2*i+1])
            counter=counter+1
            #print('counter= ', counter)
                         
#pp.pprint(price)
#pp.pprint(tvl) 
print('Число дней Price= ',len(price))
#print('Число дней TVL = ',len(tvl))
print('tvl[0]= ',tvl[0])
end = len(tvl)
print('tvl[end]= ',tvl[end-1])
for i in range(1,n+1):
    print('tvl change ',i,' = ',(tvl[end-1][i]-tvl[0][i])/tvl[0][i]*100/len(price)*365,' %  за год ')
print('Vault 1 = USDP, Vault 2 = USDN, Vault 3 = IronBank, Vault 4= sUSD ' )
for i in range(1,n+1):
    print('Profit of Vault  ',i,' = ',(price[end-1][i]-price[0][i])/price[0][i]*100/len(price)*365,' % per year')
print("Указаны проценты в пересчете на 1 год")
```
